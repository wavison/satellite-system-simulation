
## üì° **What is Eb/N‚ÇÄ?**

**Eb/N‚ÇÄ** stands for:

> **Energy per bit** divided by **Noise power spectral density**

* **Eb** ‚Üí how much energy you put into each transmitted bit
* **N‚ÇÄ** ‚Üí how much noise power exists per Hz of bandwidth

It‚Äôs a way to normalize **signal quality** so you can compare different modulations, coding rates, and symbol rates *on an equal footing*.

---

### üìè **Units & Scale**

* Measured in **decibels (dB)**:

  $$
  \text{Eb/N‚ÇÄ (dB)} = 10 \cdot \log_{10}\left(\frac{E_b}{N_0}\right)
  $$
* A **higher Eb/N‚ÇÄ** means:

  * More energy per bit relative to the noise
  * Easier for the receiver to decode the signal correctly
* A **lower Eb/N‚ÇÄ** means:

  * Bits are buried in noise
  * Higher chance of bit errors

---

### üìä **How it Relates to Modulation**

Different modulations need different **minimum Eb/N‚ÇÄ** to achieve a given bit error rate (BER), for example:

| Modulation (no coding) | Min Eb/N‚ÇÄ for BER ‚âà 10‚Åª‚Å∂ |
| ---------------------- | ------------------------ |
| BPSK                   | \~9.6 dB                 |
| QPSK                   | \~12.6 dB                |
| 8PSK                   | \~15 dB                  |
| 16APSK                 | \~16‚Äì17 dB               |
| 32APSK                 | \~18‚Äì20 dB               |

**Higher-order modulation** (e.g., 32APSK) packs more bits per symbol but is more susceptible to noise ‚Äî meaning it **requires higher Eb/N‚ÇÄ**.

---

### üõ† **How It‚Äôs Used**

* **Link budget design**: You calculate **available Eb/N‚ÇÄ** from your transmit power, antenna gains, path loss, and system noise temperature.
* Compare this to the **required Eb/N‚ÇÄ** for your chosen **MODCOD** (modulation + coding).
* The difference is your **link margin**.

Example:

* Available Eb/N‚ÇÄ = 12 dB
* Required Eb/N‚ÇÄ for QPSK 3/4 = 5.5 dB
* **Link margin** = 6.5 dB ‚Üí Good!

---

### üîç **Key Takeaways**

* **Higher is better** ‚Äî but higher means you‚Äôre wasting power *if you don‚Äôt need it*.
* Coding (like **Reed-Solomon** + **Convolutional**) reduces the required Eb/N‚ÇÄ by making the link more robust to noise.
* Trade-off:

  * **Higher-order modulation** ‚Üí More throughput, needs higher Eb/N‚ÇÄ
  * **Lower-order modulation** ‚Üí More robust, less throughput

## üì° **SNR (Signal-to-Noise Ratio)**

* Measures **signal power** compared to **noise power** in a given bandwidth:

$$
\text{SNR} = \frac{P_{\text{signal}}}{P_{\text{noise}}}
$$

* Usually expressed in **dB**:

$$
\text{SNR}_{\text{dB}} = 10 \log_{10}\left(\frac{P_{\text{signal}}}{P_{\text{noise}}}\right)
$$

* Depends on **bandwidth**: if you change the bandwidth, the SNR changes.

---

## ‚öñ **Eb/N‚ÇÄ (Energy per Bit / Noise Density)**

* Focuses on **energy per bit**, not just overall signal power.
* N‚ÇÄ is **noise power per Hz** (noise spectral density).
* It **removes the effect of bandwidth** so you can compare systems with different bit rates or modulations.

Formula linking the two:

$$
\frac{E_b}{N_0} = \frac{\text{SNR} \cdot B}{R_b}
$$

Where:

* $B$ = receiver bandwidth (Hz)
* $R_b$ = bit rate (bits/sec)
* $\text{SNR}$ = signal-to-noise ratio over bandwidth $B$

---

### üîç **Key Difference**

* **SNR** tells you ‚Äúhow strong is the signal vs noise‚Äù over a certain bandwidth.
* **Eb/N‚ÇÄ** tells you ‚Äúhow much energy is in each bit vs the noise floor‚Äù ‚Äî a fairer way to compare links with different bit rates.

---

### üìä **Example**

Imagine two systems:

| System | SNR (dB) | Bandwidth (Hz) | Bit Rate (bps) | Eb/N‚ÇÄ (dB) |
| ------ | -------- | -------------- | -------------- | ---------- |
| A      | 10       | 1 MHz          | 1 Mbps         | 10         |
| B      | 10       | 1 MHz          | 500 kbps       | 13         |

Same SNR, but **lower bit rate** means **more energy per bit**, so Eb/N‚ÇÄ is higher.
